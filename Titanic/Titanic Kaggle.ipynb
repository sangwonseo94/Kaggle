```python
import pandas as pd
```

```python
test_set = pd.read_csv('test.csv')
train_set = pd.read_csv('train.csv')
train_set
```

```python
Sex_list =[]
for i in train_set['Sex']:
    if (i == 'male'):
        Sex_list.append(1)
    else: Sex_list.append(0)
Embarked_list = []
for i in train_set['Embarked']:
    
    if ( i == 'S'): Embarked_list.append(0)
    elif( i == 'C'):  Embarked_list.append(1)
    elif( i == 'Q'):  Embarked_list.append(2)
    else:  Embarked_list.append(3)
train_set['Sex'] = Sex_list
train_set['Embarked'] = Embarked_list
```

```python
Sex_list =[]
for i in test_set['Sex']:
    if (i == 'male'):
        Sex_list.append(1)
    else: Sex_list.append(0)
Embarked_list = []
for i in test_set['Embarked']:
    
    if ( i == 'S'): Embarked_list.append(0)
    elif( i == 'C'):  Embarked_list.append(1)
    elif( i == 'Q'):  Embarked_list.append(2)
    else:  Embarked_list.append(3)
test_set['Sex'] = Sex_list
test_set['Embarked'] = Embarked_list
```

```python
train_set
```

```python
s = train_set['Age']
t = s.fillna(s.mean())
train_set['Age'] = t
```

```python
train_set = train_set[['PassengerId','Survived' , 'Pclass' , 'Sex' , 'Age' , 'SibSp', 'Parch' , 'Fare' , 'Embarked']]
test_set = test_set[['PassengerId','Pclass' , 'Sex' , 'Age' , 'SibSp', 'Parch' , 'Fare' , 'Embarked']]
```

```python
s = test_set['Age']
t = s.fillna(s.mean())
test_set['Age'] = t

s = test_set['Fare']
t = s.fillna(s.mean())
test_set['Fare'] = t
```

```python

train_label = train_set['Survived']
train_set = train_set.drop('Survived', axis=1)
# test_set = test_set.drop("PassengerId", axis=1).copy()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.utils import shuffle
```

```python
train_set, train_label = shuffle(train_set, train_label, random_state = 5)
```

```python
def train_and_test(model):
    model.fit(train_set, train_label)
    prediction = model.predict(test_set)
    accuracy = round(model.score(train_set, train_label) * 100, 2)
    print("Accuracy : ", accuracy, "%")
    return prediction
```

```python
test_set.isna().sum()
```

```python
# Logistic Regression
log_pred = train_and_test(LogisticRegression())
# SVM
svm_pred = train_and_test(SVC())
#kNN
knn_pred_4 = train_and_test(KNeighborsClassifier(n_neighbors = 4))
# Random Forest
rf_pred = train_and_test(RandomForestClassifier(n_estimators=100))
# Navie Bayes
nb_pred = train_and_test(GaussianNB())
```
